```python
df = pd.read_csv('tracks_scs.csv')
df.head(20)
```

```python
fig = plt.figure(figsize=(7, 8))

for num in np.arange(1, 2758103, 1):
#for num in np.arange(1, 448, 1):
    df_tmp = df[df.Track_Index == num]
    if df_tmp.shape[0] >= 30:
        lontmp = df_tmp.Lon
        lattmp = df_tmp.Lat
        cyctmp = df_tmp.Cyc.sum()
        if cyctmp > 0:
            plt.plot(lontmp, lattmp, color='r', linewidth=0.5)
        else:
            plt.plot(lontmp, lattmp, color='b', linewidth=0.5)
            
plt.xlim(100, 128)
plt.ylim(0, 30)
```
